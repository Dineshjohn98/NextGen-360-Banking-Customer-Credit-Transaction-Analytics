# **Stage 2 – Data Cleaning and Pre-processing**

## Handling Missing Values in Categorical and Numerical Columns

```python
# Numerical Column Inputation (Account Balance)

df['Account Balance'].fillna(df['Account Balance'].mean(),inplace=True)
```

```python
# Categorical Column Imputation (Email)

df['Email']=df['Email'].fillna("Unknown")
```

```python
df.isnull().sum()
```

### Data Type Conversions

```python
#change Loan issue date to date format
df['Loan Issue Date'] = pd.to_datetime(df['Loan Issue Date'])

```

```python
df.dtypes
```

```python
# Concatenate First and Name Last Name as Full Name

df['Full Name'] = df['First Name'] + ' ' + df['Last Name']

cols = df.columns.tolist()
last_name_idx = cols.index('Last Name')

# Move 'Full Name' to its desired position after 'Last Name'
cols.insert(last_name_idx + 1, cols.pop(cols.index('Full Name')))

df = df[cols]

df.head()
```

```python
#Drop First Name and last Name from the table

df.drop(['First Name', 'Last Name'], axis=1, inplace=True)
```

```python
df
```

```python
df.columns
```

```python
df.describe()
```

```python
# Analyse Risk Factor based on anomaly report by adding new column "Risk Factor"

df['Risk_Factor']=np.where(df['Anomaly']==1,'No Risk','Risk')
```

```python
df
```

```python
# GroupBy Card type with Minimum Payment due
# AMEX Card type highest avg min due collect from the customers

Average_Min_Due=df.groupby('Card Type')['Minimum Payment Due'].mean().sort_values(ascending=False)
Average_Min_Due
```

```python
# GroupBy Card Type with Rewards Points
# Master Card users has High avg reward points

Average_Rewards_Points=df.groupby('Card Type')['Rewards Points'].mean().sort_values(ascending=False)
Average_Rewards_Points
```

# **Stage 3 – EDA and Visualizations**

## **Univariate Analysis**

```python
# Histogram Plot
plt.figure(figsize=(10,5))
sns.histplot(df['Loan Type'],color='purple')
plt.title('Distribution of Loan Type')
plt.xlabel('Loan Type')
plt.ylabel('Frequency')
plt.show()
```

# Interpretation From Histogram plot

### **Histogram plot** defines the distribution of Loan type based on the customer frequeny it clearly emphasis that most of the customers holding the **Mortage Loan** than others respectively.

```python
 # Count Plot

plt.figure(figsize=(8,5))
colors = ['skyblue', 'lightcoral', 'lightgreen']
sns.countplot(data=df, x='Gender',palette= colors)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show
```

# Interpretation from Count Plot

### **Count Plot** defines the distribution of Gender based on the holding accounts it clearly emphasis that **Female custumers** having more accounts in bank which is really a appreciable one.  

```python
# Pie plot
plt.figure(figsize=(8,6))
df['Risk_Factor'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Fraudlent Transaction Report')
plt.ylabel('')
plt.show()
```

# Interpretation from Pie Chart

### **Pie Chart**  emphazises the Unsual or fraudlent transaction report it has been clearly defines **6%** of customers done the unsual transaction so the bank have to clearly monitor these accounts to prevent from fruadlent activities respectively.

## **Bivariate Analysis**

```python
# Box Plot
plt.figure(figsize=(8,8))
sns.boxplot(data=df, x='Gender', y='Account Balance',palette=colors)
plt.title('Account Balance by Gender')
plt.xlabel('Gender')
plt.ylabel('Account Balance')
plt.show()
```

# Interpretation from Box Plot

### **Box Plot** emphazises the relation between Gender wise account balance analysis it has been observed that the **Other category** people holding highest account balance of approximately more than 5000 which is more interesting though the female customers where in higher counts.

```python
# Bar Plot
plt.figure(figsize=(8,5))
colors = ['skyblue', 'lightcoral', 'lightgreen']
Average_Min_Due.plot(kind='bar', color=colors)
plt.title('Average Minimum Payment Due by Card Type')
plt.ylabel('Average Minimum Payment Due')
plt.show()
```

# Interpretation from Bar Plot

### **Bar Plot** describes the average minimum due by card type it has been clearly assumed that **Amex Card** holders paying higher minimum due which gradually increases bank revenue.

```python
# Violin plot
plt.figure(figsize=(15,6))
colors = ['lightgrey', 'gold', 'green'] # Highlight MasterCard with gold
sns.violinplot(data=df, x='Card Type', y='Rewards Points', palette=colors)
plt.title('Rewards Points by Card Type')
plt.xlabel('Card Type')
plt.ylabel('Rewards Points')
plt.show()
```

# Interpretation from Violin Plot

### **Violin plot** defines the reward points by card type it clearly examines the **Master Card** type has the highest reward points when compared to others.

# **Multivariate Analysis**

```python
# Scatter plot
plt.figure(figsize=(15,6))
sns.scatterplot(data=df, x='Account Balance', y='Rewards Points', hue='Gender')
plt.title('Account Balance vs Rewards Points by Gender')
plt.xlabel('Account Balance')
plt.show()
```

# Interpretation from Scatter plot

### The scatter plot shows the relationship between Account Balance and Rewards Points, categorized by Gender (Male, Female, Other). The data points are randomly distributed across the plot with no clear upward or downward trend, indicating no strong correlation between account balance and rewards points. Additionally, all three gender groups display a similar distribution, suggesting that gender does not significantly influence the relationship between account balance and rewards points.

```python
# Heat Map
plt.figure(figsize=(15,6))
numerical_cols=['Age','Rewards Points','Minimum Payment Due']
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Interpretation from Heat Map

### The correlation heatmap shows that there is no significant relationship among Age, Rewards Points, and Minimum Payment Due, as all correlation values are very close to 0. This indicates that changes in one variable have little to no impact on the others, suggesting these features are largely independent in the dataset.

```python
# Pair Plot
sns.pairplot(df[numerical_cols])
plt.show()
```

# Interpretation from Pair Plot

### The diagonal histograms in pair plot show that Age, Rewards Points, and Minimum Payment Due are fairly evenly distributed across their respective ranges.The scatter plots in pair plot indicate no strong linear relationship between any pair of variables, as the data points are randomly dispersed..
